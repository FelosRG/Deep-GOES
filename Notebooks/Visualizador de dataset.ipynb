```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt

DATASET  = "dataset.h5"

DIR_REPO = "/".join(os.getcwd().split("/")[:-1])
DIR_DATASETS = f"{DIR_REPO}/gendata/Datasets"
path_dataset = f"{DIR_DATASETS}/{DATASET}"

print("Variables disponibles:")
with h5py.File(path_dataset,"r") as file:
    keys = file.keys()
    for key in keys:
        print(key," : ",file[key].shape)
        
# Algunas funciones Ãºtiles

def extraer_variable(variable):
    with h5py.File(path_dataset,"r") as file:
        array = file[variable][:]
    return array


def plot_histograma(array,bins=30):
    array = array.reshape(-1)
    n,_,_ = plt.hist(array,bins=bins,density=True,stacked=False)
    plt.show()
    return n
    
def plot_ejemplos(array,vmin,vmax):
    num_t = array.shape[0]
    num_l = array.shape[1]
    array = array.reshape((num_t*num_l,*array.shape[2:]))
    array = array[index]
    fig , ax = plt.subplots(1,num_ejemplos)
    for i in range(num_ejemplos):
        ax[i].imshow(array[i],vmin=vmin,vmax=vmax)
        ax[i].axis("off")
    plt.show()
```

```python
num_t = 432
num_l = 900
num_ejemplos = 7
index = np.random.randint(0,num_t*num_l,num_ejemplos)
```

### Variable: Banda 4

```python
array = extraer_variable("4")
```

```python
_ = plot_histograma(array)
```

```python
vmin,vmax = 0,20
plot_ejemplos(array,vmin,vmax)
```
